```python
from langgraph.graph import END,StateGraph,START
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import threading
from pydantic import BaseModel,Field
from typing import Optional,Annotated
from langgraph.checkpoint.memory import InMemorySaver
from langgraph.graph.message import add_messages
from langchain_core.messages import BaseMessage, AIMessage, SystemMessage, HumanMessage,AIMessageChunk
from langchain_core.runnables import RunnableConfig
import os
load_dotenv()
```

```python
class chatState(BaseModel):
    messages:Annotated[list[BaseMessage],add_messages]= Field(description="user message")

```

```python


model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

def llmchatNode(State: chatState):
    messages = State.messages
    res = model.invoke(messages)
    return {"messages": [AIMessage(content=res.content)]}

graph = StateGraph(chatState)
graph.add_node("llmChat_node", llmchatNode)
graph.add_edge(START, "llmChat_node")
graph.add_edge("llmChat_node", END)

chatbot = graph.compile()
while True:
    user_input= input("write ")
    res=chatbot.invoke({"messages":[HumanMessage(content=user_input)]})
    # print(res["messages"][-1]>)
   
```
